```python
datelines = pd.to_datetime(retweets['created_at'])
dates =[datelines[i].date() for i in range(len(datelines))]


from datetime import timedelta, date

def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days)):
        yield start_date + timedelta(n)
        
def missingDates(dates):
    start_date = min(dates)
    end_date = max(dates)
    unique_dates = {}
    for dt in dates:
        if not dt in unique_dates: 
            unique_dates[dt] = 1
    missing = []
    for dt in daterange(start_date, end_date):
        if not dt in unique_dates:
            missing.append(dt)
    return missing        
        
missings =missingDates(dates)     
print(len(missings))
#for d in missings:
#    print(d.strftime("%Y-%m-%d"))
        
```
